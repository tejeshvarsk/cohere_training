```python
from cohere.responses.chat import StreamEvent

for event in co.chat("What is an LLM?", stream=True):
    if event.event_type == StreamEvent.TEXT_GENERATION:
      print(event.text)
    elif event.event_type == StreamEvent.STREAM_END:
      print(event.finish_reason)
```

```python
from cohere.responses.classify import Example
```

```python
examples = [Example("This is a bad example","positive"),
            Example("This is a worst example","positive"),
            
            Example("This is a neutral example","negative"),
            Example("This is a common example","negative"),
            
            Example("This is a good example","neutral"),
           Example("This is a best example","neutral")
           ]
inputs=["this is not an unseen example", "this is an unseen example"]
```

```python
classifications=co.classify(model='embed-english-v3.0', examples=examples, inputs=inputs).classifications
```
